```python
import flowutils
import matplotlib.pyplot as plt
import flowio

# Load the FCS file
fcs_file_path = "./fcsdaata.fcs"
data = flowio.FlowData(fcs_file_path)

# Get the channel names
channel_names = data.channels

# Specify the channels to plot
x_channel = "CD22"
y_channel = "CD19"

# Get the data for the specified channels
x_data = data[:, x_channel]
y_data = data[:, y_channel]

# Create a scatter plot
plt.scatter(x_data, y_data)
plt.xlabel(x_channel)
plt.ylabel(y_channel)
plt.title("FCS Scatter Plot")
plt.show()
```
